```python
import pandas as pd
import numpy as np
import os
```

```python
raw_data_path=os.path.join(os.path.pardir,'data','raw')
train_data_path=os.path.join(raw_data_path,'train.csv')
test_data_path=os.path.join(raw_data_path,'test.csv')
```

```python
train_df=pd.read_csv(train_data_path, engine='python', index_col='PassengerId')
test_df=pd.read_csv(test_data_path,engine='python', index_col='PassengerId')
```
